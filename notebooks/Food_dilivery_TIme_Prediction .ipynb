```python
# let's convert decimal points values into minute format 
import pandas as pd
import numpy as np

# Function to convert decimal to time
def convert_decimal_to_time(decimal):
    try:
        # Convert decimal to float
        decimal = float(decimal)
        # Calculate hours and minutes
        hours = int(decimal * 24)
        minutes = int((decimal * 24 * 60) % 60)
        return f"{hours:02}:{minutes:02}"
    except ValueError:
        # Return the original string if it is not a decimal
        return decimal
```

```python
# Applying Function to convert decimal to time
data_copy['Time_Orderd'] = data_copy['Time_Orderd'].apply(convert_decimal_to_time)
```

```python
# Split the time into hours and minutes
data_copy[['orderd_hours', 'orderd_minutes']] = data_copy['Time_Orderd'].str.split(':', expand=True)
```

```python
def fill_null(data_copy):
    for i in data_copy.columns:
        if data_copy[i].isnull().sum() > 0:
            if data_copy[i].dtype == 'object':  # Check if dtype is string/object
                data_copy[i].fillna(data_copy[i].mode()[0], inplace=True)
            else:
                data_copy[i].fillna(data_copy[i].mean(), inplace=True)
    
    return data_copy
```

```python
# Now Let's Move towards model training 
# Form Model training we have to convert our categorical columns into numrecical columns

data_copy.head()
```

```python
whether_categories = ['Fog' , 'Stormy' , 'Sandstorms' , 'Windy' , 'Cloudy' , 'Sunny']
traffic_condition = ['Jam' , 'High' , 'Medium' , 'Low']
meal_type = ['Snack' , 'Meal' , 'Drinks' , 'Buffet']
vehicle_type = ['motorcycle' , 'scooter' , 'electric_scooter' , 'bicycle']
festival_condition = ['No' , 'Yes']
city_type = ['Metropolitian' , 'Urban' , 'Semi-Urban']
```

```python
# Dividing data into independent and dependent features 
X = data_copy.drop('Delivery_time',axis=1)
Y = data_copy['Delivery_time']
```

```python
# Getting categorical and numerical columns 
numerical_col = [col for col in X.columns if X[col].dtype != "object" ]
categorical_col = [col for col in X.columns if X[col].dtype == "object" ]
```

```python
from sklearn.impute import SimpleImputer # Handling missing values 
from sklearn.preprocessing import StandardScaler # Handling feature scalling
from sklearn.preprocessing import OrdinalEncoder # Ordinal Encoding --> Categorical features 

# Pipelines 
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer # --> To combine two pipelines 
```

```python
# Assuming you have defined 'numerical_col' and 'categorical_col' with the names of numerical and categorical columns, respectively.

## Numerical Pipeline
num_pipeline=Pipeline(
    steps=[
    ('imputer',SimpleImputer(strategy='median')),
    ('scaler',StandardScaler())

    ]

)

# Categorigal Pipeline
cat_pipeline=Pipeline(
    steps=[
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('ordinalencoder',OrdinalEncoder(categories=[whether_categories,traffic_condition,meal_type,vehicle_type,festival_condition,city_type])),
    ('scaler',StandardScaler())
    ]

)

preprocessor=ColumnTransformer([
('num_pipeline',num_pipeline,numerical_col),
('cat_pipeline',cat_pipeline,categorical_col)
])
```

```python
# Train and test split to get train and test data to train model 
from sklearn.model_selection import train_test_split 
X_train , X_test , y_train , y_test = train_test_split(X,Y , test_size = 0.33 , random_state=42)
```

```python
import numpy as np
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square
```

```python
## Train multiple models

models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'Elasticnet':ElasticNet()
}
trained_model_list=[]
model_list=[]
r2_list=[]

for i in range(len(list(models))):
    model=list(models.values())[i]
    
    # Here the models getting trained 
    model.fit(New_X_train,y_train)

    #Make Predictions always based on test data
    y_pred=model.predict(New_X_test)

    # By obtaining the predicted values, you can compare them 
    # with the actual target values (y_test) to evaluate the model's
    # performance and assess how well it predicts the 
    # target variable on unseen data.

    mae, rmse, r2_square=evaluate_model(y_test,y_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])

    print('Model Training Performance')
    print("RMSE:",rmse)
    print("MAE:",mae)
    print("R2 score",r2_square*100)

    r2_list.append(r2_square)
    
    print('='*35)
    print('\n')
```
